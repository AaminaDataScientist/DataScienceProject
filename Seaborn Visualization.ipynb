# Seaborn Visualization

```python
import pandas as pd
import matplotlib.pyplot as  plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
sns.load_dataset('tips')
```

```python
tips=sns.load_dataset('tips')
print(tips.head())
sns.violinplot(x="total_bill", data=tips)
plt.show()
```

```python
sns.distplot(tips["tip"], bins=20)
```

```python
sns.distplot(tips["total_bill"], bins=20, kde=False)
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip")
```

```python
sns.rugplot(data=tips, x='tip')
```

```python
sns.rugplot(data=tips, x='total_bill', y='tip')
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette="coolwarm")
```

```python
sns.pairplot(tips)
```

```python
sns.catplot(x='day', y="total_bill", kind="box", data=tips, palette="coolwarm")
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", data=tips, kind="box")
plt.title("Customer Vists")
```
